```python
X = data_house['loyer']
y = data_house['surface']
```

```python
plt.figure(figsize=(15,5))
sns.scatterplot(X,y, color='darkblue')
```

```python
n_samples = data_house.shape[0]
```

```python
# Define learning rate
LR = 0.001
theta_0 = 0
theta_1 = 0
```

```python
def partial_derivative(theta_0, theta_1):
    deriv_theta_0 = 0
    deriv_theta_1 = 1
    for i in range(0, n_samples):
        deriv_theta_0 += theta_0 + (theta_1 * X[i]) - y[i]
        deriv_theta_1 += (theta_0 + (theta_1 * X[i]) - y[i]) * X[i]
    
    deriv_theta_0 = (1/n_samples) * deriv_theta_0
    deriv_theta_1 = (1/n_samples) * deriv_theta_0
    
    return deriv_theta_0, deriv_theta_1
```

```python
def get_new_theta(prev_theta_0, prev_theta_1):
    
    deriv_theta_0, deriv_theta_1 = partial_derivative(prev_theta_0, prev_theta_1)
    new_theta_0 = prev_theta_0 - (LR * deriv_theta_0)
    new_theta_1 = prev_theta_1 - (LR * deriv_theta_1)
    
    return new_theta_0, new_theta_1
```

```python
def cost_function(theta_0, theta_1):
    cost = 0
    for i in range(0,n_samples):
        error = ((theta_0 + theta_1 * X[i])- y[i])**2
        cost += error
        
    print("Cost : ", (1/(2*n_samples))*cost)
    
    #return (1/(2*n_samples))*cost
```

```python
def grad_descent():
    tmp_0 = 0
    tmp_1 = 0
    
    for i in range(0,1000):
        tmp_0, tmp_1 = get_new_theta(tmp_0, tmp_1)
        cost_function(tmp_0, tmp_1)


    return tmp_0, tmp_1
```

```python
theta_0 , theta_1 = grad_descent()
pred = []

for x in X:
    predic = theta_0 + theta_1 * x
    pred.append(predic)
```

```python
plt.plot(X,pred,color='red')
sns.scatterplot(X, y, color='darkblue')
```
